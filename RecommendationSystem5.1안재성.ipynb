## 1.5 Model-based CF 알고리즘


###    ALS 교대 최소 제곱법
            행렬 R을 분해하는 가장 간단한 방법 
          
          
\\(R\simeq PQ^T = \hat{R}\\)

여기서 P(N*K) 는 특징 공간에서의 새로운 유저 행렬, Q(M*K)는 특징공간에서의 아이템의 투영Protection이다. 따라서 낮은 차원으로 문제를 축소해 정의한 비용함수 J를 최소화하는 P와 Q를 찾을 수 있다.


\\(J=min \underset{ij}\Sigma \underset{ij}e^2 = \underset{p,q}min \underset{ij}\Sigma Mc_{ij} \begin{pmatrix} r_{ij}- \sum_{k=1}^K p_{ik}q_{kj}\end{pmatrix}^2 + \frac{\lambda}{2}\begin{pmatrix} \left|p_i\right|^2 + \left|{q_j}^T\right|^2 \end{pmatrix}\\)

\\(\lambda\\)는 overfitting방지.

이후에 J를 사용자 벡터 pi와 아이템 벡터 qj에 대하여 미분해 0으로 하면 pi와 qj에 관한 식이 나온다.
이 식을 최소 제곱 알고리즘(least square algorithm)(최소자승법)으로 직접 푼다.

## 1.5 Model-based CF 알고리즘 SVD
<br/>
1. m-step  : \\(\hat{R}=SVD(\hat{R})\\) 실행<br/><br/>

2. e-step  : \\(\underset{ij}{\hat{r}}= r_{ij} \\) if (\\(\underset{ij}r\\) is filled by the user) <br/>\\(=\underset{ij}{\hat{r}}\\) else (missing data) 
\\(A = U\Sigma V^T\\)

![Imgur](http://i.imgur.com/MKqhtji.png)

특징 1: ALS보다 느리나, 정확도는 높다<br/>

SVD 분해는 memory-based CF 에서 축소된 공간(행렬 U나 \\(V^T\\))에서 사용자나 아이템을 비교하기위해 사용될 수 있고, 이때의 평점은 원래의 유틸리티 행렬에서 추출됨. (K-NN 방식을 이용한 SVD)
